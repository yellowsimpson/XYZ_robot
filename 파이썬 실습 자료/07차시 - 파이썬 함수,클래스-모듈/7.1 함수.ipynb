```python
scores = [85, 90, 70, 65, 80] # 학생들의 점수를 리스트로 저장

total = 0 # 총점 변수 초기화
for score in scores:
    total += score # 각 점수를 총점에 더함

average = total / len(scores) # 총점을 학생 수로 나눠 평균 계산

print("학생들의 평균 점수는", average, "입니다.")
```

### 7.1.3 함수 입력값(매개변수) 응용 : *args

*args : def 를 이용하여 함수를 정의할 때 매개변수를 *매개변수명 와 같은 형태로 입력하면 개수가 정해지지 않은 입력값을 매개변수에 tuple 형태로 저장할 수 있음

```python
def test(*args):
	print('args:', args)
	print('type:', type(args))
 
test(1, 2, 3, 4, 5) # 출력값 : args: (1, 2, 3, 4, 5) / type: <class 'tuple'>
 
def test(a, *args):
	print('a:', a)
	print('args:', args)
	print('type:', type(args))
 
test(0, 1, 2, 3, 4, 5) # 출력값 : a: 0 / args: (1, 2, 3, 4, 5) / type: <class 'tuple'>
```

### 7.1.5 재귀함수

return으로 자기 자신의 함수를 다시 호출하여 값을 가져오는 것

```python
def fibonacci(n):
    print(n,end=',')
    if n == 0 or n == 1: # 초기값 설정
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2) # 재귀 호출

print(fibonacci(5)) # 8
#print(fibonacci(10)) # 89
```

```python
def factorial(n):
    print(n,end='*')
    if n==1:
        return 1
    else:
        return n * factorial(n-1)

factorial(5)
```

### 2. 로컬변수 (Local Variable):

로컬변수는 함수나 블록 내부에서 선언된 변수입니다.

해당 함수나 블록 내부에서만 접근 가능하며, 함수가 종료되면 사라집니다.

로컬변수는 동일한 이름의 전역변수를 가려버릴 수 있습니다.

```python
k = 3

def add(a):
    global k

    a = a + 1
    return a

print(a)

a = add(2)

print(a)
```

```python
class Person:
    def __init__(self,first_name,second_name): #클래스를 처음 만들때 호출하는 메소드
        self.first_name = first_name
        self.second_name = second_name
        
    def print_name(self,hi):
        print(hi, ' ', self.second_name,self.first_name,'님')

    def hollo(self,word):
        self.word = word

leelanhee = Person('란희','이') #클래스를 인스턴스화...=>클래스의 생성자를 호출
leelanhee.print_name('안녕하세요') # 클래스의 메소드를 사용하기위해서는 먼저 그 클래스를 인스턴스화 한다.
```
