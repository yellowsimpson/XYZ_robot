### 4.1.5 if ~ else문

두가지 조건 즉 판단을 했을 때 참이 아닌 경우에 else문을 사용

if-else 문은 주어진 조건의 참 부분과 거짓 부분을 모두 실행하는 데 사용됩니다. 조건이 참이면 if 블록 코드가 실행되고 조건이 거짓이면 else 블록 코드가 실행됩니다.


```python
money = True 

if money == True:
    print("택시를 타고 간다")
else:
    print("걸어서 간다")
    
# True, False (대소문 구분)
```

```python
money = input(' 가진 돈이 얼마에요?')

money = int(money)
r = (money >= 4000)

if not r:
    print("택시를 타고 간다")
else:
    print("걸어서 간다")
    
```

### 4.1.6 중첩 if문

지금까지 if를 한 번만 혹은 else를 사용하여 2가지 조건에서의 단순한 조건문을 사용했습니다. 

이보다 더 복잡한 경우를 프로그래밍을 하다 보면 if를 여러 번 사용하는 복잡한 조건도 자주 발생합니다. 이번에는 if를 여러 번 사용하는 중첩 if 조건문을 배워보겠습니다. 

다음은 변수의 값이 10 이상이면 '10 이상입니다.'를 출력한 뒤 15이면 '15입니다.', 20이면 '20입니다.'를 출력하는 구문입니다. 

이러한 것을 중첩 조건(if)문이라하고 이는 3개 이상 무한하게 가능합니다.

```python
x = 15
 
if x >= 10:
     print('10 이상입니다.') 
        
     if x == 15:  
         print('15입니다.')
            
     if x == 20:# 
         print('20입니다.')
     else:
         print("15나 20이 아닙니다..");
```

주의할 점은 들여쓰기 탭 즉 4칸의 의미를 이해하고 적용하여야합니다.

```python
x = 15
 
if x <= 10:
     print('10 이하입니다.') 
else:  
    print('10 초과입니다.')
          
    if x == 15:
        print('15입니다.')
    else:    
         print('15가 아닙니다.')
```

```python
num = 30

if num >= 20:
    if num > 31:
        print("30보다 크다.")
    else:
        print("30보다 작거나 같다.")
else:
    print("20보다 작다.")
```

### 4.1.7 elif문

위에서와 같이 조건을 지속적으로 판별하되 하나의 조건에만 해당되는 경우에 쉽게 하는 방법.

위에서 방법의 단점을 생각해봅시다.

elif문 : if ~ else를 줄임

```python

score = 45 #int(input("점수를 입력 ==> "))

if score >= 90:
    print('A', end='')
elif score >= 80:
    print('B', end='')
elif score >= 70:
    print('C', end='')
elif score >= 60:
    print('D', end='')
elif score >= 50:
    print('E', end='')
else :
    print('F', end='')

    
print("학점입니다.")

#여러개의 구간으로 판별을 해야하는 경우 크거나 또는 작은 순서대로 문법을 구성
```

### 4.1.9 if ~ else문 응용 (조건부 표현식)

if ~ else문에서 if조건문과 else에 충족할 경우 수행할 코드가 둘 다 1줄이라면 이 if ~ else문을 1줄로 표현할 수 있습니다.

```python
num = 30
even_or_odd = 'even' if num % 2 == 0 else 'odd'
print(even_or_odd)
```

변수 = 조건이 **참**인 경우의 **값** **if** 조건문 **else** 조건이 **거짓**일 경우의 **값**

```python
print("자기가 갖고있는 돈") 
x=int(input()) 

if x>10000:
    print("택시타기")
elif x>=1500 and x<=10000:
    print("버스타기")
else:
    print("걸어가기")
```

```python
money = int(input("돈 얼마 있어요? : "))

if money >= 4000:
    print("택시를 타고 가세요")
elif 1000 <= money < 4000:
    print("버스나 지하철을 타고 가세요")
else:
    print("걸어서 가세요")
```

```python
# x = dollar on hand
#transportation = walk, take bus, grab taxi

def transportation(x):
    if x < 1:
        print("walk")
    if x >= 1 and x < 2:
        print("take bus")
    if x >= 2:
        print("grab taxi")
            
transportation(3)
```

### 어떤 숫자를 입력받으면 짝수인지, 홀수인지 판단해서 프린트해주는 프로그램.

```python
def odd_even(n):
    if n % 2 == 0:
        print(f'{n} is even number(짝수)')
    else:
        print(f'{n} is odd number(홀수)')

        
num = 'q'


while not num.isdigit():
    num = input("숫자를 입력하시오")
    if not num.isdigit():
        print(f"{num} 은 숫자가 아닙니다")
        continue
        
odd_even(int(num))
```

```python

num = int(input("숫자를 입력하시오.")) #input 입력을 받으면 그 리턴값은 문자열(str)

if num % 2 == 0:
    print("짝수입니다.")
else :
    print("홀수입니다.")
```

### 돌발 문제


자신이 근무한 시간을 입력받는다.

주당 40시간 이상 근무한 경우 시급의 1.5배를 받는다.

그래서 주급을 계산하라.

1. 시급은 결정되어 있다.(정하시기 바랍니다.)
2. 주간 근무한 시간을 입력받는다.
3. 40시간 이상근무했는지, 그렇지 않은지 판다.
4. 그것에 따라서 주급을 계산하고 프린트하라...


```python
work_hour = int(input("근무 시간 입력하세요."))
hourly_pay = 9620

if work_hour >= 40:
    print(hourly_pay * 1.5 * work_hour)
else:
    print(hourly_pay * work_hour)
```

60시간 이상은 2배를 줍니다... 수정하세요...

```python
def payCal(weekly_time):
    pay_time = 9620
    
    if 40 <= weekly_time < 60:
        return pay_time * weekly_time * 1.5
    elif 60 <= weekly_time:
        return pay_time * weekly_time * 2
    else:
        return pay_time * weekly_time

def payCal_2(weekly_time):
    pay_time = 9620
    
    if 60 <= weekly_time:
        return pay_time * weekly_time * 2
    elif 40 <= weekly_time:
        return pay_time * weekly_time * 1.5
    else:
        return pay_time * weekly_time
    
def validator():
    while True:
        try:
            work_time = int(input("일한 시간은?"))  #int()에서 exception이 발생할 수 있다.
        except:
            print("시간은 정수입니다")
            continue
        break
    return work_time

print(f'{payCal(validator())} 원을 주급으로 지급한다')
```

금액을 ,가 있는 형태로 바꿔주세요.

```python
money = validator()
weekly_pay = payCal(money)

weekly_pay = format(weekly_pay,',')

print(f'{weekly_pay} 원을 주급으로 지급한다') ###가독성을 높여라...
```

```python
l = [1,2,3]

def isin(x):
    for i in l:
        if x == i:
            return True
        
    return False

tf = isin(0)

if tf:  # if tf == True:
    print('IS')
else:
    print('IS NOT')
```

```python
def is_same(a,b):
    s = (a == b)
    if s==True:
        print(a,'and',b,' is same')
    else:
        print(a,'and',b,' is not same')  


a = input('first number:')
b = input('second number:')
a = int(a)
b = int(b)

is_same(a,b)
```

```python
def between(a):
    if (a >= 10) and (a <= 20):
        return True
    else:
        return False
    
i = input('number between 10 ~ 20:')    
i = int(i)
r = between(i)
if r==True:
    print('good')
else:
    print('bad')
```
