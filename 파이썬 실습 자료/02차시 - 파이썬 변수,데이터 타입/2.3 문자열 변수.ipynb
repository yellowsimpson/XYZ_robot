### 2.3.3 문자열 접근

파이썬에서 문자열은 문자열의 각 문자마다 0부터 1씩 증가하는 번호를 매겨놓습니다.<br>
이걸 인덱스라 부르고 이를 통해 문자열에 접근할 수 있습니다.

```python
# [i]를 사용해서 문자열에서 인덱스 i에 있는 문자 하나에 접근하기 
say  = "나는 너를 좋아해 라고 영수가 말했다."
print(say)
print(say[0],say[1])
```

문자열 끝에서부터 셀 수도 있습니다.

```python
say  = "\"나는 너를 좋아해.\" 라고 영수가 말했다."

print(say[-2])
```

문자열을 뒤에서 부터 셀 때에는 처음엔 -0이 와야하는 것이 아닌가? 라고 잠깐 생각해볼 수 있지만<br>
0과 -0은 동일하므로 같은 값을 출력합니다.<br>
그렇기에 뒤에서부터 문자열을 셀 경우에는 -1이 시작입니다.

```python
str  = "문자열을 뒤에서부터 세봅시다."

print(str[0])

print(str[-0])

print(str[-1])
```
