### 1.2.1 변수(Variable) 시작하기

````python
a = 10
# 10이라는 숫자를 a라는 곳에 담어줘, 이때 a는 10을 대신해서 사용하고자 합니다.
# 이제 a에 잘 저장되었는지 프린트로 확인해봅니다.
print(a)
````

파이썬 변수의 특징은 다른 프로그래밍 언어인 C나 JAVA에서는 변수를 만들 때 자료형의 타입을 직접 지정해야 하지만 <br>파이썬은 변수에 저장된 값을 스스로 판단하여 자료형의 타입을 지정하기 때문에 <br>타입을 일치시켜야하는 딱딱한 규칙을 지킬 필요가 없습니다.

### 1.2.4 카피하기

1. [:] 이용<br>
첫 번째 방법으로는 다음과 같이 리스트 전체를 가리키는 [:]을 사용해서 복사하는 겁니다.

````python
a = [1, 2, 3]
b = a[:]
````

2. copy 모듈 이용<br>
두 번째 방법은 copy 모듈을 사용하는 겁니다. 

````python
from copy import copy
a = [1, 2, 3]
b = copy(a)
````
