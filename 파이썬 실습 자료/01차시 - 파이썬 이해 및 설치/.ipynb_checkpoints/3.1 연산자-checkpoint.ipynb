## 1.6 연산자

### 1.6.1 산술 연산자

#### 변수 a의 값 10에 20 증가시키려면 어떻게 해야 할까요?
#### a + 20처럼 20을 더하면 30이 나오지만 a의 값을 다시 출력해보면 10이 나옵니다

```python
a = 10
a + 20
#print(a)
```

```python
print( a + b )    # 더하기
print( a - b )    # 빼기
print( a * b )    # 곱하기
print( a / b )    # 나누기
print( a // b )   # 나눈 몫
print( a % b )    # 나눈 나머지
```

```python
x = 2
y = 3
sum = x + y
print (sum)
```

```python
x = 8
y = 3
mod = x % y
print (mod)
```

산술 연산자는 기본 우선 순위를 따릅니다. 둘 이상의 연산자가 사용되는 경우 다음 순서에 따라 실행됩니다.

### 1.6.2 비교 연산자

프로그래밍에서 값을 다른 값과 비교하려는 경우가 종종 있습니다. 이를 위해 비교 연산자를 사용합니다.

Python에는 다음과 같은 6개의 비교 연산자가 있습니다.

미만 ( <)
보다 작거나 같음( <=)
보다 큼 ( >)
크거나 같음 ( >=)
같음 ( ==)
같지 않음 ( !=)
이러한 비교 연산자는 두 값을 비교하고 부울 값( True또는 )을 반환합니다 False.

그리고 이러한 비교 연산자를 사용하여 숫자 와 문자열 을 모두 비교할 수 있습니다 .

```python
10 < 20
```

```python
20 >= 10
```

### 1.6.3 할당 연산자

할당 연산자는 변수에 값을 할당하는 데 사용됩니다.

```python
x = 5

x += 3

print(x)
```

### 1.6.4 논리 연산자

논리 연산자는 조건문을 결합하는 데 사용됩니다.
논리연산자에는 and, or, not 이 있는데, and 는 양쪽의 값이 모두 참인 경우만 참이 되고, or 는 어느 한쪽만 참이면 참이 됩니다.
not 은 참이면 거짓으로 거짓이면 참이 됩니다. 

```python
x = 5

print(x > 3 and x < 10)
```

```python
x = 5

print(not(x > 3 and x < 10))
```

### 1.6.5 멤버쉽 연산자

멤버쉽연산자에는 in, not in 이 있는데, 이는 좌측 Operand가 우측 컬렉션에 속해 있는지 아닌지를 체크합니다. 파이썬 멤버쉽 연산자는 문자열, 리스트 또는 튜플과 같은 시퀀스 멤버십을 테스트하는데 사용됩니다.

```python
x = ["apple", "banana"]

print("banana" in x)

# returns True because a sequence with the value "banana" is in the list
```
