문자열 메소드들

text[i]	문자열에서 인덱스 i에 있는 문자 하나에 접근합니다. 

text.count(a)	문자열에서 a가 등장하는 횟수를 반환합니다.

text.find(a)	문자열에서 a가 등장하는 곳의 인덱스를 반환합니다. a가 여러 번 등장한다면 가장 첫 번째의 인덱스를 반환합니다. a가 없으면 -1을 반환합니다.

text.index(a)	find와 유사하지만 찾을 수 없는 경우에는 에러를 일으킵니다.


```python
text = "문자열에서 a가 등장하는 곳의 인덱스를 반환합니다"

print(f"{text=}")
print(f"{text[3]=}")
print(f"{text.count('a')=}")
print(f"{text.count('nothing')=}")

print(f"{text.find('build')=}")

print(f"{text.find('등장')=}")

print(f"{text.index('a가')=}")
```

```python


text = "Do you wanna build a snowman?"
print(f"{text.partition('wanna')=}")

print(f"{text.split(' ')=}")
print(f"{text.split('nothing')=}")

print(f"{text.replace('build', 'throw')=}")
```

```python
print(f"{'  하이    '.strip()=}")
```

```python
l = ['1','하이','파워','국어']
print(f"{l=}")
print(f"{''.join(l)=}")
print(f"{', '.join(l)=}")
```

### 4.1.5 인덱싱

인덱싱(Indexing)이란 무엇인가를 "가리킨다"는 의미이고, 슬라이싱(Slicing)은 무엇인가를 "잘라낸다"는 의미이다. 
이런 의미를 생각하면서 다음 내용을 살펴보자.
아래 그림은 실제 메모리에 저장되는 방식이고 이를 인덱싱을 이용 접근하는 것이다.

```python
hello = 'Hello, world!'
hello[len(hello) - 1]    # 마지막(인덱스 12) 문자 출력
```

문자열 인덱스를 알면 이후에 배울 리스트를 이해하는데 도움이 됩니다.

### 4.1.6 index()함수

이 메서드는 찾은 하위 문자열 의 첫 번째 위치 를 반환합니다.

index()아래에는 String 함수 의 작동을 보여주는 예가 있습니다.

```python
ab = 'We can do whatever we want'
sample = ab.index('w', 5, 20)
print(sample)
```

```python
#다음을 실행해보자
str1="Baaceian hwwers"
a=str1.index('e')
print(a)
#여러 값이 있는 경우 해당 값이 처음 나타나는 인덱스를 반환합니다.
str2="Zebra Crossings"
b=str2.index('r',5,9)
print(b)

str3="Hoopla"
print(str3.index('e')) # 없으면 에러...
```
