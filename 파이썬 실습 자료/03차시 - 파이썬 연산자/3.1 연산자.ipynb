# 3.1 연산자

### 3.1.1 산술 연산자

#### 변수 a의 값 10에 20 증가시키려면 어떻게 해야 할까요?
#### a + 20처럼 20을 더하면 30이 나오지만 a의 값을 다시 출력해보면 10이 나옵니다

```python
a = 10
b = 20
c = a + 20

print(c)
```

```python
print( a + b )    # 더하기
print( a - b )    # 빼기
print( a * b )    # 곱하기
print( a / b )    # 나누기
print( a // b )   # 나눈 몫
print( a % b )    # 나눈 나머지
```

```python
x = 2
y = 3
sum = x + y
print (sum)
```

```python
x = 8
y = 3
mod = x % y
print (mod)
```

산술 연산자는 기본 우선 순위를 따릅니다. 둘 이상의 연산자가 사용되는 경우 다음 순서에 따라 실행됩니다.

### 3.1.3 비교 연산자

프로그래밍에서 값을 다른 값과 비교하려는 경우가 종종 있습니다. 이를 위해 비교 연산자를 사용합니다.

Python에는 다음과 같은 6개의 비교 연산자가 있습니다.

미만 ( <)
보다 작거나 같음( <=)
보다 큼 ( >)
크거나 같음 ( >=)
같음 ( ==)
같지 않음 ( !=)
이러한 비교 연산자는 두 값을 비교하고 부울 값( True또는 )을 반환합니다 False.

그리고 이러한 비교 연산자를 사용하여 숫자와 문자열을 모두 비교할 수 있습니다.
문자열은 사전 순서로 크기를 비교합니다.

```python
10 > 20
```

```python
20 >= 10
```

```python
num = 6
print(5 < num < 10) # 논리 연산자 없이 값의 범위 구하기
```

```python
a = '김민수'
b = '이은채' #사전 순서
print(a < b)
```

```python
3 >= 1
```

### 3.1.4 논리 연산자

논리 연산자는 조건문을 결합하는 데 사용됩니다.
논리연산자에는 and, or, not 이 있는데, and 는 양쪽의 값이 모두 참인 경우만 참이 되고, or 는 어느 한쪽만 참이면 참이 됩니다.
not 은 참이면 거짓으로 거짓이면 참이 됩니다. 

```python
x = 1

print(x > 3 or x > 10)
```

```python
x = 5

print(not(x > 3 and x < 10))
```
