### 8.2.1 파일 입출력 시작

값을 "입력"받을 때는 사용자가 직접 입력하는 방식을 사용했고 "출력"할 때는 모니터 화면에 결괏값을 출력하는 방식을 사용했다. 

open()파일 객체 를 반환하며 두 개의 위치 인수와 하나의 키워드 인수와 함께 가장 일반적으로 사용됩니다.

```python
f = open('live.txt','w')

f.write("0 라인")
f.write("1 라인")
f.write("2 라인")
f.write("3 라인")
f.write("4 라인")

f.close()
```

```python
f = open('live_lines.txt','w')

f.write("0 라인\n")
f.write("1 라인\n")
f.write("2 라인\n")
f.write("3 라인\n")
f.write("4 라인\n")

f.close()
```

```python
f = open('live.txt','r')

t = f.read()

print(t)

f.close()
```

### 파일 열기 open 옵션들

파일 읽기(쓰기)순서

1. 파일의 경로 찾기
상대 경로와 절대 경로를 모두 사용하여 파일을 읽을 수 있습니다. 경로는 디스크에 있는 파일의 위치입니다. 절대 경로 에는 파일을 찾는 데 필요한 전체 디렉터리 목록이 포함됩니다. 파일 경로는 OS에따라 다를 수 있습니다.

2. 읽기 모드에서 파일 열기
파일 을 열려면 파일 경로와 액세스 모드 를 open()함수에 전달합니다. 액세스 모드는 읽기 또는 쓰기와 같이 파일에서 수행하려는 작업을 지정합니다. 오픈 모드를 잘 확인하고 열기를 합니다. 위 옵션 참조

3. 파일에서 콘텐츠를 읽습니다.
파일을 열면 메소드 read()를 사용하여 파일의 모든 텍스트 또는 내용을 읽을 수 있습니다.  readline()파일을 한 줄씩 읽기를 사용하거나, 모든 줄을 읽기 위해 readlines()를 사용할 수도 있습니다.
4. 읽기 작업 완료 후 파일 닫기
파일 작업을 완료한 후 파일이 제대로 닫히는지 확인해야 합니다. close() 함수는 파일을 닫는 데 사용 합니다. 파일을 닫지 않으면 작업한 내용이 사라집니다.

### 8.2.2 JSON 구조화된 데이터 저장

```python
import json

x = [1, 'simple', 'list']

json.dumps(x)
```

```python
#dumps()라는 함수 의 또 다른 변형은 dump()단순히 객체를 텍스트 파일 로 직렬화합니다 . 
#따라서 쓰기를 위해 열려 f있는 텍스트 파일 개체인 경우 다음과 같이 할 수 있습니다.

f = open('./files/jsonwork.txt', 'w+')

json.dump(x, f)

f.close()
```

```python
# 개체를 다시 디코딩하려면 if f가 읽기용으로 열린 이진 파일 또는 텍스트 파일 개체인 경우:

f = open('./files/jsonwork.txt', 'r+')

k = json.load(f)

k
```

### 8.2.5 with으로 파일 읽기,쓰기

```python
# 파일을 열고 닫을 때 필수인 close()를 호출하지 않고 with으로 해결

with open('./files/live.txt', encoding="utf-8") as f:
    read_data = f.read()

read_data
```

```python
with open('with.txt','w') as f:
    f.write('good')
```

```python
import os

# folder path
dir_path = './/'

# list to store files
res = []

# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        res.append(path)
print(res)
```

### 8.2.7  파일 입출 시 경로

(1) 절대 경로 

파일을 읽거나 쓸 때 

특정 경로로 지정해준다.

ex) 'C:/Users/wow/Desktop/data/data.txt'



(2) 상대 경로

'./data/data.txt'와 같이 프로그램이 실행되는 폴더 기준으로 

상대적인 주소가 지정되는 것. 다른 pc에서 작업하더라도

같은 workspace 라면./data 전의 주소는 상대적이기 때문에 편리.

```python
import csv

def save_contact():
    id_name = input("아이디 입력: ")
    name = input("이름 입력: ")
    phone_number = input("전화번호 입력: ")

    # CSV 파일 열기 또는 생성
    with open('members.csv', mode='a+', newline='') as file:
        writer = csv.writer(file)

        # 헤더가 없는 경우에만 쓰기
        if file.tell() == 0:
            writer.writerow(['id', 'name', 'phone'])

        # 데이터 쓰기
        writer.writerow([id_name, name, phone_number])

    print("연락처가 성공적으로 저장되었습니다.")

# 프로그램 실행
save_contact()
```
