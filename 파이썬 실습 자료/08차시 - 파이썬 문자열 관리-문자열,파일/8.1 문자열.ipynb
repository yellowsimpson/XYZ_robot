### 8.1.3 문자열 인덱싱

인덱싱(Indexing)이란 무엇인가를 "가리킨다"는 의미입니다.

아래 그림은 실제 메모리에 저장되는 방식이고 이를 인덱싱를 이용해 접근합니다.

문자열은 인덱스로 접근하여 문자 하나를 가져올 수 있습니다. 

문자열을 자유롭게 인덱스를 사용해서 자를 수 있습니다. 

단 문자열을 중간에 인덱스로 바꿀 수는 없습니다.

(문자열 인덱스를 잘 이해하면 이후에 배울 리스트를 이해하는데 도움이 됩니다.)

```python
string = 'Good Morning!'

print(string[0])
print(string[3])
print(string[4])
print(string[-1])
```

```python
string = 'Good Morning!'
print(string[1],end=':')
print(string[2],end=':')
```

```python
string = 'Good Morning!'
for c in string:
    print(c,end='')
```

```python
s = 'hello'
w = ' world'

s * 3
```

### 8.1.4 문자열 슬라이싱

슬라이싱(Slicing)이란 무엇인가를 ‘인덱싱 정보를 활용해서 잘라낸다’라는 의미입니다.

사용하는 방법은 문자열[시작번호:끝번호]에서 시작 번호와 끝 번호를 생략하면 문자열의 처음부터 끝까지 값을 가져올 수 있습니다.

```python
string = 'Good Morning!'

print(string[2:7]) # 끝자리는 미만으로 한다.
print(string[0:4])
print(string[5:len(string)])

print(string)
```

```python
string = 'Good Morning!'
print(string[4:-1])
```

```python
a = "Hello world"

a[0:4]
```

이렇게 되는 이유는 a[시작 번호:끝 번호]를 지정하면 끝 번호에 해당하는 것은 포함되지 않습니다.

이 수식을 만족하는 a는 a[0], a[1], a[2], a[3]입니다. 

따라서 a[0:4]은 ‘Hell’이고 a[0:5]는 ‘Hello’가 되는 것입니다.

### 8.1.5 문자열 덧셈

파이썬에서 문자열과 문자열을 연결할 때는 문자열을 마치 숫자를 더하는 것처럼 더하면 됩니다.

### 8.1.7 문자열 루프 (Loop)

문자열를 이터레이트 (iterate)하기 위해 아래와 같이 for 루프를 사용할 수 있습니다.

```python
# for루프는 목록, 문자열 등과 같은 구조를 반복하는 데 사용됩니다. 문자열은 본질적으로 반복 가능합니다.
# 즉, 문자열을 반복하면 각 문자가 출력으로 제공됩니다.

string = '문자열을 반복하면'

for c in string:
  print(c,end=' ')
```

```python
# while루프는 주어진 조건이 True가 될 때까지 주어진 명령문 세트에 대한 for 루프처럼 사용됩니다.
# 문자열을 반복하기 위해len()함수를 사용하여 문자열의 길이를 제공합니다.

Str_value = "문자열을 반복하기"

i = 0; 
while(i < len(Str_value)):
    
    print(Str_value[i],end=' ')
    
    i=i+1
```

### 8.1.10 문자열 출력 형식 지정

```python
"[%20s]" % "*"   # [와 ] 사이에 20칸의 공백이 있다.
```

```python
"[%-20s]" % "A"  # 20칸의 공백의 앞쪽에 A를 인쇄한다.
```

```python
"[%20d]" % 123  # 20칸의 공백의 뒷쪽에 123을 인쇄한다.
```

```python
"[%20d]" % 123  # 20칸의 공백의 뒷쪽에 123을 인쇄한다.
```

```python
x = 1 / 3.0  # 값은 0.3333333....
"%.5f" % x   # 소숫점 아래 5자리까지만 인쇄한다.
```

### 8.1.12 트리만들기 - 이런게 알고리즘입니다.

```python
line = int(input("Tree 의 높이를 입력하세요(5~30) : "))

for x in range(1, line * 2, 2):
    print((" " * ( (line * 2 - 1 - x) // 2 )) + ("*" * x))
    
for y in range(1, 4):
    print(" " * (line-2) + "***")
    
```

```python
# 과제 다이아몬드 만들기

line = int(input("Diamond 의 길이를 입력하세요(2~30) : "))

for x in range(1, line * 2, 2):
    print((" " * ( (line * 2 - 1 - x) // 2 )) + ("*" * x))

for y in range(line * 2-3, 0, -2):
    print((" " * ( (line * 2 - 1 - y) // 2 )) + "*" * y)
        
```
