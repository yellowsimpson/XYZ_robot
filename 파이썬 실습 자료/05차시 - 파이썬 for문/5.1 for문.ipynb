#### for문은 몇번 반복할지를 어느정도 알고 그 값이 정의있는 경우....

* #### for 루프는 콜렉션 자료들 (즉, 목록, 튜플, 사전, 집합 또는 문자열)과 같은 시퀀스한 정보를 처리하고 프로세스를 반복하는 데 사용됩니다.

* #### for 루프를 사용하면 리스트, 튜플, 세트 등의 각 항목에 대해 한 번씩 일련의 명령문을 실행할 수 있습니다.

### 5.1.3 반목문 제어 - break

break 문을 사용하면 모든 항목을 반복하는 과정 중에 루프를 중지할 수 있습니다. <br>
if 조건문과 함께 특정 조건을 충족시키면 멈추는 식으로 자주 사용됩니다.

```python
fruits = ["사과", "배", "파인에플","감"]

for i in range(10):
    print(i)
    
    for x in fruits:
        if x == "파인에플":
        break
        
    print(x)

print("끝")
```

### 미션
['Siya', 'Tiya', 'Cito', 'Daksh', 'Riya', 'Cito']로 구성된 리스트(my_list)는 루프를 사용하여 반복됩니다. 우리는 목록에서 'Cito'라는 이름을 검색하려고 합니다.
for ~ loop 내부에서 if 조건은 my_list의 요소를 'Cito'라는 이름과 비교합니다. 조건이 참(true)이면 중단문이 실행되고 반복문이 종료됩니다.

```python
my_list = ['Siya', 'Tiya', 'Cito', 'Daksh', 'Riya', 'Cito'] 

for i in range(len(my_list)):
    print(my_list[i])
    if my_list[i] == 'Cito':
        print('Found the name Cito')
        break
        print('After break statement')

print('Loop is Terminated')
```

```python
my_list = ['Siya', 'Tiya', 'Cito', 'Daksh', 'Riya', 'Cito'] 

for i in my_list:
    print(i)
    if i == 'Cito':
        print('Found the name Cito')
        break
        print('After break statement')

print('Loop is Terminated')
```

```python
my_list = ['Siya', 'Tiya', 'Cito', 'Daksh', 'Riya', 'Cito'] 
i = 0

while True:
    print(my_list[i])
    if (my_list[i] == 'Cito'):
        print('Found the name Cito')
        break
        print('After break statement')
    i += 1

print('After while-loop exit')
```
